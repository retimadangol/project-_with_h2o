### 2. Subset the data to people less than $200k/annual to avoid huge outliers

```python
loan_df_2 = loan_df[loan_df['annual_inc'] < 200000]
```

### 3. Tokenize the words in the column

```python
loan_df_2['emp_title'] = loan_df_2['emp_title'].ascharacter()
```

### 4. Transform the original dataset with the word2vec model

```python
tokenized_df = loan_df_2['emp_title'].tokenize(" ")
tokenized_df.head(5)
```

```python
tokenized_df[tokenized_df['C1']==''].shape
```

### 5. Building word to vec model from the tokenized words

```python
w2v_model = H2OWord2vecEstimator()
w2v_model.train(training_frame=tokenized_df)
```

```python
# just double checking everything is cool
w2v_model.find_synonyms("Executive", count = 5)
```

```python
# just double checking everything is cool
w2v_model.find_synonyms("Manager", count = 5)
```

### 6. Transform original dataset with the model

```python
title_vec = w2v_model.transform(tokenized_df, aggregate_method = "AVERAGE")
```

### 7. Cbind the transformed vectors with the response column "annual_inc"

```python
title_vec_annual_inc = loan_df_2['annual_inc'].cbind(title_vec)
```

```python
title_vec_annual_inc.head(5)
```

```python
title_vec_annual_inc.shape
```

```python
loan_df_2.shape
```

### 8. Build a regression model to predict annual income based on job title

```python
from h2o.estimators.glm import H2OGeneralizedLinearEstimator
predictors = title_vec_annual_inc.columns[1:]
target = 'annual_inc'
train, valid, test = title_vec_annual_inc.split_frame([0.7, 0.15], seed=1234)
```

```python
glm_model = H2OGeneralizedLinearEstimator(family= "gaussian", 
                                          lambda_ = 0, 
                                          remove_collinear_columns = True,
                                          compute_p_values = True)
glm_model.train(predictors, 
                target , 
                training_frame = train,
                validation_frame = valid)
```
